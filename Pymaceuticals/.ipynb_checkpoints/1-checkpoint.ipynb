```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
clinical = pd.read_csv(clinical_trial_data_to_load)
mouse.head(2)
```

```python
clinical.head(2)
```

```python

# Combine the data into a single dataset
data = pd.merge(mouse, clinical, on = "Mouse ID", how = "outer" )

# Display the data table for preview
#data_copy = data (need format?)
#data_copy['Tumor Volume (mm3)'] = data_copy['Tumor Volume (mm3)'].map("{:.1f}".format)
data.head(2)
```

# 1st above

## 2 Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint  (by group?)not getting the same as the example
tumor_mean= data.groupby(["Drug","Timepoint"])['Tumor Volume (mm3)'].mean()
# # Convert to DataFrame

tumor_mean_df = pd.DataFrame(tumor_mean)
tumor_mean_df.head()

tumor_mean_df['Tumor Volume (mm3)'].dtype
# # Preview DataFrame

tumor = tumor_mean_df.reset_index()
tumor.head()

# tumor_mean_df['Tumor Volume (mm3)'].dtype
# tumor['Tumor Volume (mm3)'].dtype
```

# 2nd above


```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_group =  data.groupby(["Drug","Timepoint"])['Tumor Volume (mm3)'].sem()


# Convert to DataFrame
tumor_sem_df = pd.DataFrame(sem_group)


# Preview DataFrame

tumor_sem_df.head()

view_tumor_sem = tumor_sem_df.reset_index()
view_tumor_sem.head()
```

```python
tumor_sem2 = view_tumor_sem.pivot(columns ="Drug", values = "Tumor Volume (mm3)", index = "Timepoint")
tumor_sem2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames  
tumor2 = tumor.pivot(columns ="Drug", values = "Tumor Volume (mm3)", index = "Timepoint")


# # Preview that Reformatting worked 
tumor2.head()
```

```python
# means =[tumor2['Capomulin'], tumor2['Infubinol'],tumor2['Ketapril'],tumor2['Placebo']]
# sems =[tumor_sem2['Capomulin'],tumor_sem2['Infubinol'],tumor_sem2['Ketapril'],tumor_sem2['Placebo']]
# Generate the Plot (with Error Bars Capomulin,Infubinol, Ketapril, and Placebo
# fig, ax = plt.subplots()
x_axis = tumor2.index

# ax.errorbar(x_axis, means, sems)

plt.errorbar (x_axis,tumor2['Capomulin'],tumor_sem2['Capomulin'],marker ='o',color = 'red', label = "Capomulin",linestyle ='dashed')

plt.errorbar(x_axis, tumor2['Infubinol'], tumor_sem2['Infubinol'],marker ='^',color = 'blue',label = 'Infubinol',linestyle ='dashed')
plt.errorbar(x_axis, tumor2['Ketapril'], tumor_sem2['Ketapril'],marker = 's',color = 'green',label = 'Ketapril',linestyle='dashed')
plt.errorbar(x_axis, tumor2['Placebo'], tumor_sem2['Placebo'],marker ='d',color = 'purple',label = 'Placebo',linestyle='dashed')

#plt.xlim(-0.5, max(tumor2.index)+0.5)
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title('Tumor Response to Treatment')
plt.grid()


# Save the Fig
#plt.savefig("../Images/Tumor Response to Treatment.png")
# How to make the line to "---"



```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
Met_mean_group = data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()
Met_mean_group.head()
# Convert to DataFrame
Met_mean_df =pd.DataFrame(Met_mean_group)
Met_mean = Met_mean_df.reset_index()
# Preview DataFrame
Met_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
Met_sem_group = data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem()

# Convert to DataFrame
Met_sem_df = pd.DataFrame(Met_sem_group)
Met_sem = Met_sem_df.reset_index()

Met_sem2 = Met_sem.pivot(columns ="Drug", values = "Metastatic Sites", index = "Timepoint")



# Preview DataFrame
Met_sem2.head(3)
```

```python
# Minor Data Munging to Re-Format the Data Frames
Met_mean2 = Met_mean.pivot(columns ="Drug", values = "Metastatic Sites", index = "Timepoint")
# Preview that Reformatting worked
Met_mean2.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = Met_mean2.index
plt.errorbar(x_axis, Met_mean2['Capomulin'],Met_sem2['Capomulin'], marker = 'o',color = "red",label = 'Capomulin',linestyle='dashed')
plt.errorbar(x_axis, Met_mean2['Infubinol'], Met_sem2['Infubinol'],marker ='^',color ='blue',label ='Infubinol',linestyle='dashed')
plt.errorbar(x_axis, Met_mean2['Ketapril'], Met_sem2['Ketapril'], marker = 's', color = 'green',label ='Ketapril',linestyle ='dashed')
plt.errorbar(x_axis, Met_mean2['Placebo'], Met_sem2['Placebo'], marker = 'd',color ='purple',label = 'Placebo',linestyle ='dashed')

plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Met. Sites)")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title('Metastatic Spread During Treatment')
plt.grid()

# Save the Figure
plt.savefig("../Images/Metastatic Spread During Treatment.png")
# Show the Figure
#plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_group = data.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
# Convert to DataFrame
count_df = pd.DataFrame(count_group)

# Preview DataFrame
count_mouse = count_df.reset_index()
count_mouse.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mouse2 = count_mouse.pivot(columns = "Drug", values = 'Mouse ID', index= 'Timepoint')

# Preview the Data Frame
count_mouse2.head(2)
```

```python
total = count_mouse2.iloc[0,:]
survive_rate = count_mouse2/total*100
survive_rate.head()
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = survive_rate.index
plt. plot(x_axis, survive_rate['Capomulin'],marker = 'o',color='red',label='Capomulin',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Infubinol'],marker = '^',color='blue', label ='Infubinol',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Ketapril'],marker='s',color='green',label='Ketapril',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Placebo'],marker='d',color='purple',label='Placebo',linestyle ='dashed')

plt.xlabel('Time(Days)')
plt.ylabel('Survive Rate(%)')
plt.title('Survival During Treatment')
plt.legend(loc = 'best', fontsize ="small" , fancybox = True)
plt.grid()

# Save the Figure
plt.savefig( '../Images/Survival During Treatment')

# Show the Figure
```
