```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
clinical = pd.read_csv(clinical_trial_data_to_load)
# Combine the data into a single dataset
data=pd.merge(mouse,clinical, on ='Mouse ID',how='outer')

# Display the data table for preview
data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
data_group = data.groupby(['Drug','Timepoint'])
tumor_mean=data_group['Tumor Volume (mm3)'].mean()
# Convert to DataFrame
tumor_mean_df = pd.DataFrame(tumor_mean)
# Preview DataFrame
tumor_mean_df.head()
tumor_mean2=tumor_mean_df.reset_index()
tumor_mean2.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem=data_group['Tumor Volume (mm3)'].sem()
# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_sem)
tumor_sem2 = tumor_sem_df.reset_index()
# Preview DataFrame
tumor_sem2.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_mean3=tumor_mean2.pivot(columns='Drug',index='Timepoint',values='Tumor Volume (mm3)')
tumor_sem3=tumor_sem2.pivot(columns='Drug',index='Timepoint',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_mean3.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = tumor_mean3.index
plt.errorbar(x_axis,tumor_mean3['Capomulin'],tumor_sem3['Capomulin'],marker ='o',color = 'red', label = "Capomulin",linestyle ='dashed')
plt.errorbar(x_axis, tumor_mean3['Infubinol'], tumor_sem3['Infubinol'],marker ='^',color = 'blue',label = 'Infubinol',linestyle ='dashed')
plt.errorbar(x_axis, tumor_mean3['Ketapril'], tumor_sem3['Ketapril'],marker = 's',color = 'green',label = 'Ketapril',linestyle='dashed')
plt.errorbar(x_axis, tumor_mean3['Placebo'], tumor_sem3['Placebo'],marker ='d',color = 'purple',label = 'Placebo',linestyle='dashed')
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title('Tumor Response to Treatment')
plt.grid()

# Save the Figure
plt.savefig('images/Tumor Response to Treatment')
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Generate the Plot (with Error Bars)
x_axis = Met_mean3.index
plt.errorbar(x_axis, Met_mean3['Capomulin'],Met_sem3['Capomulin'], marker = 'o',color = "red",label = 'Capomulin',linestyle='dashed')
plt.errorbar(x_axis, Met_mean3['Infubinol'], Met_sem3['Infubinol'],marker ='^',color ='blue',label ='Infubinol',linestyle='dashed')
plt.errorbar(x_axis, Met_mean3['Ketapril'], Met_sem3['Ketapril'], marker = 's', color = 'green',label ='Ketapril',linestyle ='dashed')
plt.errorbar(x_axis, Met_mean3['Placebo'], Met_sem3['Placebo'], marker = 'd',color ='purple',label = 'Placebo',linestyle ='dashed')
plt.xlabel("Treatment Duration(Days)")
plt.ylabel("Met. Sites)")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title('Metastatic Spread During Treatment')
plt.grid()

# Save the Figure
plt.savefig('images/Metastatic Spread During Treatment')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_group = data.groupby(["Drug", "Timepoint"])["Mouse ID"].count()
# Convert to DataFrame
count_df = pd.DataFrame(count_group)
# Preview DataFrame
count_mouse = count_df.reset_index()
count_mouse.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
count_mouse2 = count_mouse.pivot(columns = "Drug", values = 'Mouse ID', index= 'Timepoint')
# Preview the Data Frame
count_mouse2.head(2)
```

```python
total = count_mouse2.iloc[0,:]
survive_rate = count_mouse2/total*100
survive_rate.head()
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = survive_rate.index
plt.plot(x_axis, survive_rate['Capomulin'],marker = 'o',color='red',label='Capomulin',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Infubinol'],marker = '^',color='blue', label ='Infubinol',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Ketapril'],marker='s',color='green',label='Ketapril',linestyle ='dashed')
plt.plot(x_axis, survive_rate['Placebo'],marker='d',color='purple',label='Placebo',linestyle ='dashed')

plt.xlabel('Time(Days)')
plt.ylabel('Survive Rate(%)')
plt.title('Survival During Treatment')
plt.legend(loc = 'best', fontsize ="small" , fancybox = True)
plt.grid()

# Save the Figure
plt.savefig( 'images/Survival During Treatment')
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
pct_change = (tumor_mean3.iloc[-1,:]-tumor_mean3.iloc[0,:])/tumor_mean3.iloc[0,:]*100

# Display the data to confirm
pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
tuple_changes = (pct_change["Capomulin"], 
               pct_change["Infubinol"], 
               pct_change["Ketapril"], 
               pct_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
x_axis = np.arange(len(tuple_changes))  
width = 1

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(x_axis + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([min(tuple_changes)-10,max(tuple_changes)+10])
ax.grid(True)

# Use functions to label the percentages of changes
# plt.text(x, y, s, (optional fontdict=None, withdash=<deprecated parameter>, **kwargs)
def fail_label(Fail):
    for i in Fail:
        height = i.get_height()
        ax.text(i.get_x() + i.get_width()/2.,1,
                '%d%%' % int(height))

def pass_label(Pass):
    for n in Pass:
        height = n.get_height()
        ax.text(n.get_x() + n.get_width()/2., -18,
                '-%d%% ' % int(height))

# Call functions to implement the function calls
Pass = ax.bar(x_axis[0], tuple_changes[0],width,color='green')
Fail = ax.bar(x_axis[1:], tuple_changes[1:],width, color='red')
pass_label(Pass)
fail_label(Fail)

# Save the Figure
fig.savefig("images/pct changes")

# Show the Figure
fig.show()
```
